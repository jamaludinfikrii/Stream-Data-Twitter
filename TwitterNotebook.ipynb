```python
import pandas as pd
import json
```

```python
tweetList = []
```

```python
with open('machinelearning.json') as json_file:
    data = json.load(json_file)
    
    for item in data:
        hashtags = ''
        for hashtag in item['entities']['hashtags']:
            hashtags += hashtag['text'] + ' ';
        tweetList.append([item['id'],
                          item['created_at'],
                          item['text'],
                          item['user']['id'],
                          item['user']['name'],
                          item['user']['screen_name'],
                          hashtags,
                          item['source']])

dfStream = pd.DataFrame(tweetList, columns = ['id','createdAt','text','userId','userName','ScreenName','hastag','source'])
dfStream.head()
```
